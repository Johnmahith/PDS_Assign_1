# 2) Perform 5 data visualization tasks on the student performance dataset given in the link below (create 5 different visualizations). Explain what kind analysis has become easier with each of the visualizations

```python
# uploading the dataset
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/StudentsPerformance.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
#finding null values
df.isnull().sum()
```

```python
#creating the cleaned data file
clean_data_path = '/content/drive/MyDrive/clean_student_data.csv'
df.to_csv(clean_data_path, index=False)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Set the plots
sns.set_style("whitegrid")

# 1. Gender Distribution in Scores
plt.figure(figsize=(18, 5))
for i, subject in enumerate(['math score', 'reading score', 'writing score'], start=1):
    plt.subplot(1, 3, i)
    sns.histplot(data=df, x=subject, hue="gender", element="step", stat="density", common_norm=False)
    plt.title(f'{subject.title()} by Gender')
plt.tight_layout()
plt.show()
```

```python
# 2. Average Scores by Race/Ethnicity
plt.figure(figsize=(18, 5))
for i, subject in enumerate(['math score', 'reading score', 'writing score'], start=1):
    plt.subplot(1, 3, i)
    sns.barplot(x="race/ethnicity", y=subject, data=df, ci=None, palette="coolwarm")
    plt.title(f'Average {subject.title()} by Race/Ethnicity')
plt.tight_layout()
plt.show()
```

```python
# 3. Parental Education Level and Student Scores
plt.figure(figsize=(18, 5))
for i, subject in enumerate(['math score', 'reading score', 'writing score'], start=1):
    plt.subplot(1, 3, i)
    sns.boxplot(x=subject, y="parental level of education", data=df, palette="Set3")
    plt.title(f'{subject.title()} by Parental Education')
plt.tight_layout()
plt.show()
```

```python
# 4. Effect of Lunch Type on Scores
plt.figure(figsize=(18, 5))
for i, subject in enumerate(['math score', 'reading score', 'writing score'], start=1):
    plt.subplot(1, 3, i)
    sns.violinplot(x="lunch", y=subject, data=df, palette="Set2")
    plt.title(f'{subject.title()} by Lunch Type')
plt.tight_layout()
plt.show()
```

```python

# 5. Test Preparation Course Impact
plt.figure(figsize=(18, 5))
for i, subject in enumerate(['math score', 'reading score', 'writing score'], start=1):
    plt.subplot(1, 3, i)
    sns.barplot(x="test preparation course", y=subject, data=df, ci=None, palette="Paired")
    plt.title(f'{subject.title()} by Test Preparation Course')
plt.tight_layout()
plt.show()
```


- Among the five visualizations created, the Test Preparation Course Impact visualization provides the most straightforward analysis. This grouped bar chart directly compares the average scores in math, reading, and writing for students who completed a test preparation course against those who did not. It clearly illustrates the difference in performance across the three subjects based on whether or not a student engaged in test preparation. This visualization simplifies the understanding of the direct impact of test preparation courses on student performance, making it the easiest to interpret for assessing the value of these courses.
